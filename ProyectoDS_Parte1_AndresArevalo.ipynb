A continuacin, describo detalladamente cada uno de los campos incluidos en la base de datos proporcionada:

Descripcin de Campos en la Base de Datos:

**1. REC_ID:** Identificador nico de cada registro en la base de datos, facilitando la referencia y el anlisis de cada lote de caf.

**2. Species:** Especie botnica de los granos de caf, como Arabica o Robusta, que son conocidas por sus distintas cualidades y sabores.

**3. Continent.of.Origin:** Continente de origen del lote de caf, proporcionando un contexto geogrfico para el anlisis.

**4. Country.of.Origin:** Pas de origen del lote de caf, esencial para el anlisis regional y la comparacin de calidad.

**5. Harvest.Year:** Ao de cosecha del lote de caf, importante para entender la frescura y posibles variaciones anuales en la calidad.

**6. Expiration:** Fecha de caducidad asignada al lote de caf, indicativa de la vida til y calidad mantenida.

**7. Variety:** Variedad especfica o tipo de planta de caf de la que se cosechan los granos, influenciando directamente las caractersticas del caf.

**8. Color:** Color observado del grano de caf crudo, que puede ser azul, verde o mixto, sugiriendo diferencias en el procesamiento o calidad.
Processing.Method: Mtodo utilizado para procesar los granos de caf despus de la cosecha, que puede incluir mtodos como lavado, natural o honey.

**9. Aroma:** Fragancia o aroma del caf, crucial en la evaluacin sensorial del producto.

**10. Flavor:** Evala el sabor del caf, incluyendo dulzura, amargura, acidez y otras notas de sabor.

**11. Aftertaste:** Gusto persistente que queda en la boca despus de consumir el caf.

**12. Acidity:** Acidez del caf, refiriendo a la vivacidad o brillo del sabor.
Body: Espesor o viscosidad del caf en la boca, importante para la textura general.

**13. Balance:** Cmo los diferentes componentes del sabor del caf se complementan entre s.

**14. Uniformity:** Consistencia del caf de taza a taza, indicando la calidad y consistencia de la produccin.

**15. Clean.Cup:** Si el caf est libre de sabores extraos o defectos, como la acidez, el moho o la rancidez.

**16. Sweetness:** Dulzura percibida en el paladar al probar el caf.

**17. Moisture:** Contenido de humedad de los granos de caf, generalmente expresado como un porcentaje.

**18. Quakers:** Presencia de granos quakers, que son granos inmaduros o defectuosos que no se tuestan adecuadamente.

**19. Category.One.Defects:** Nmero total de defectos de primer nivel, como granos negros o agrios.

**20. Category.Two.Defects:** Nmero total de defectos de segundo nivel, que incluyen defectos ms severos como granos mohosos, daados por insectos o perforados.

**21. Rec_Cnt (Grouped by data sets only):** Cuenta total de registros para cada uno de los contenedores de datos de ubicacin respectivos.

## Histograma de Dulzura por Pas (Colombia vs Otros)

Este grfico nos ayudar a visualizar la distribucin de la dulzura del caf en Colombia en comparacin con otros pases

```python
# Filtrar datos para Colombia y otros pases
colombia = Data[Data['Country.of.Origin'] == 'Colombia']
others = Data[Data['Country.of.Origin'] != 'Colombia']

# Histograma
plt.figure(figsize=(10, 6))
plt.hist(colombia['Sweetness'], alpha=0.5, label='Colombia', bins=15)
plt.hist(others['Sweetness'], alpha=0.5, label='Otros pases', bins=15)
plt.xlabel('Dulzura')
plt.ylabel('Frecuencia')
plt.title('Distribucin de la Dulzura del Caf: Colombia vs Otros Pases')
plt.legend()
plt.show()
```

**Conclusiones: **

* La mayora de las muestras de caf, tanto de Colombia como de otros pases, estn concentradas en un valor alto de dulzura, especficamente en la escala de 10, lo que indica una percepcin de dulzura muy fuerte en estas muestras.

* Tanto en Colombia como en otros pases, hay una predominancia de muestras con alta dulzura

## Diagrama de Caja de Uniformidad por Continente
