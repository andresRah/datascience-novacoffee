A continuación, describo detalladamente cada uno de los campos incluidos en la base de datos proporcionada:

Descripción de Campos en la Base de Datos:

**1. REC_ID:** Identificador único de cada registro en la base de datos, facilitando la referencia y el análisis de cada lote de café.

**2. Species:** Especie botánica de los granos de café, como Arabica o Robusta, que son conocidas por sus distintas cualidades y sabores.

**3. Continent.of.Origin:** Continente de origen del lote de café, proporcionando un contexto geográfico para el análisis.

**4. Country.of.Origin:** País de origen del lote de café, esencial para el análisis regional y la comparación de calidad.

**5. Harvest.Year:** Año de cosecha del lote de café, importante para entender la frescura y posibles variaciones anuales en la calidad.

**6. Expiration:** Fecha de caducidad asignada al lote de café, indicativa de la vida útil y calidad mantenida.

**7. Variety:** Variedad específica o tipo de planta de café de la que se cosechan los granos, influenciando directamente las características del café.

**8. Color:** Color observado del grano de café crudo, que puede ser azul, verde o mixto, sugiriendo diferencias en el procesamiento o calidad.
Processing.Method: Método utilizado para procesar los granos de café después de la cosecha, que puede incluir métodos como lavado, natural o honey.

**9. Aroma:** Fragancia o aroma del café, crucial en la evaluación sensorial del producto.

**10. Flavor:** Evalúa el sabor del café, incluyendo dulzura, amargura, acidez y otras notas de sabor.

**11. Aftertaste:** Gusto persistente que queda en la boca después de consumir el café.

**12. Acidity:** Acidez del café, refiriendo a la vivacidad o brillo del sabor.
Body: Espesor o viscosidad del café en la boca, importante para la textura general.

**13. Balance:** Cómo los diferentes componentes del sabor del café se complementan entre sí.

**14. Uniformity:** Consistencia del café de taza a taza, indicando la calidad y consistencia de la producción.

**15. Clean.Cup:** Si el café está libre de sabores extraños o defectos, como la acidez, el moho o la rancidez.

**16. Sweetness:** Dulzura percibida en el paladar al probar el café.

**17. Moisture:** Contenido de humedad de los granos de café, generalmente expresado como un porcentaje.

**18. Quakers:** Presencia de granos quakers, que son granos inmaduros o defectuosos que no se tuestan adecuadamente.

**19. Category.One.Defects:** Número total de defectos de primer nivel, como granos negros o agrios.

**20. Category.Two.Defects:** Número total de defectos de segundo nivel, que incluyen defectos más severos como granos mohosos, dañados por insectos o perforados.

**21. Rec_Cnt (Grouped by data sets only):** Cuenta total de registros para cada uno de los contenedores de datos de ubicación respectivos.

## Histograma de Dulzura por País (Colombia vs Otros)

Este gráfico nos ayudará a visualizar la distribución de la dulzura del café en Colombia en comparación con otros países

```python
# Filtrar datos para Colombia y otros países
colombia = Data[Data['Country.of.Origin'] == 'Colombia']
others = Data[Data['Country.of.Origin'] != 'Colombia']

# Histograma
plt.figure(figsize=(10, 6))
plt.hist(colombia['Sweetness'], alpha=0.5, label='Colombia', bins=15)
plt.hist(others['Sweetness'], alpha=0.5, label='Otros países', bins=15)
plt.xlabel('Dulzura')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Dulzura del Café: Colombia vs Otros Países')
plt.legend()
plt.show()
```

**Conclusiones: **

* La mayoría de las muestras de café, tanto de Colombia como de otros países, están concentradas en un valor alto de dulzura, específicamente en la escala de 10, lo que indica una percepción de dulzura muy fuerte en estas muestras.

* Tanto en Colombia como en otros países, hay una predominancia de muestras con alta dulzura

## Diagrama de Caja de Uniformidad por Continente

Comparación de la uniformidad del café en diferentes continentes

```python
plt.figure(figsize=(12, 8))
plt.boxplot([Data[Data['Continent.of.Origin'] == continent]['Uniformity'] for continent in Data['Continent.of.Origin'].unique()],
            labels=Data['Continent.of.Origin'].unique())
plt.ylabel('Uniformity')
plt.title('Uniformidad del Café por Continente')
plt.show()
```

**Conclusiones:**

* La mayoría de los continentes presentan una alta uniformidad en la calidad del café, con muchas observaciones cercanas al máximo (10), lo que indica una consistencia muy alta en las muestras de café.

* **Outliers:** Existen algunos valores atípicos, especialmente en America del Norte y Asia, donde algunos lotes de café tienen valores de uniformidad significativamente más bajos. Esto podría indicar problemas de consistencia en ciertas áreas o en ciertos lotes específicos dentro de esos continentes.

* América del Sur muestra una de las uniformidades más altas y consistentes. Esto sugiere que Colombia, como parte de América del Sur, probablemente también muestra alta uniformidad en la calidad del café. Esto apoya la hipótesis de que Colombia puede ser un lugar adecuado para producir un café de alta calidad y uniforme.

## Scatter Plot de Dulzura vs Uniformidad

* La dulzura, aunque varía, **no muestra una correlación clara con la uniformidad**, lo que indica que otros factores además de la ubicación geográfica podrían influir en la dulzura del café. * Correlaciones Altas entre Aroma, Sabor y Cuerpo:
  
1.   Aroma y Sabor (0.81): Esta es una correlación muy alta, lo que indica que la percepción del aroma está fuertemente ligada a la percepción del sabor en el café.
2.   Sabor y Cuerpo (0.74): También es una correlación fuerte, sugiriendo que los cafés con un sabor más intenso tienden a tener un cuerpo más pleno.
