##Agregar en Nulos "falta agregar datos" con polars :)

```python
for column in penguins.columns:
    penguins[column] = penguins[column].fillna("falta agregar datos")


penguins.head()
```

```python
pl_penguins1=pl.DataFrame()

for column in pl_penguins.columns:
    pl_penguins=pl_penguins.with_columns(pl.col(column).fill_null('falta agregar datos'))

pl_penguins
```

```python
import plotly.express as px
```

```python
# Gr√°fico de barras para contar las especies
# Crear el gr√°fico de l√≠neas con Plotly Express

fig = px.bar(
    penguins,
    x='species',
    #y='species',
    color='species',  # Color diferenciado por cada acci√≥n
    title='Distribuci√≥n de especies de ping√ºinos'
)

fig.update_layout(
    height=600,  # Ajustar la altura de la figura
    width=1000,  # Ajustar el ancho de la figura
    xaxis_title='species',
    yaxis_title='',
)

fig.show()
```

```python
# Crear un gr√°fico circular (gr√°fico de pastel) para contar las especies

especies_contadas = penguins['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(especies_contadas, labels=especies_contadas.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribuci√≥n de especies de ping√ºinos")
plt.show()
```

```python
# Gr√°fico de dispersi√≥n para visualizar la relaci√≥n entre variables num√©ricas
penguins = sns.load_dataset('penguins')
sns.pairplot(penguins, hue = 'species')
plt.show()
```
