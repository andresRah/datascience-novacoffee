##Agregar en Nulos "falta agregar datos" con polars :)

```python
for column in penguins.columns:
    penguins[column] = penguins[column].fillna("falta agregar datos")


penguins.head()
```

```python
pl_penguins1=pl.DataFrame()

for column in pl_penguins.columns:
    pl_penguins=pl_penguins.with_columns(pl.col(column).fill_null('falta agregar datos'))

pl_penguins
```

```python
import plotly.express as px
```

```python
# Gráfico de barras para contar las especies
# Crear el gráfico de líneas con Plotly Express

fig = px.bar(
    penguins,
    x='species',
    #y='species',
    color='species',  # Color diferenciado por cada acción
    title='Distribución de especies de pingüinos'
)

fig.update_layout(
    height=600,  # Ajustar la altura de la figura
    width=1000,  # Ajustar el ancho de la figura
    xaxis_title='species',
    yaxis_title='',
)

fig.show()
```

```python
# Crear un gráfico circular (gráfico de pastel) para contar las especies

especies_contadas = penguins['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(especies_contadas, labels=especies_contadas.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribución de especies de pingüinos")
plt.show()
```

```python
# Gráfico de dispersión para visualizar la relación entre variables numéricas
penguins = sns.load_dataset('penguins')
sns.pairplot(penguins, hue = 'species')
plt.show()
```

```python
#Encontrar el rango intercuartil (IQR) de la masa corporal de los pingüinos para cada especie.
#es una medida de dispersión estadística que describe la extensión de los valores en el rango medio de un conjunto de datos.

iqr_by_species = penguins.groupby('species')['body_mass_g'].quantile(0.75) - df_penguins.groupby('species')['body_mass_g'].quantile(0.25)

# Mostrar los resultados
print(iqr_by_species)
```
